```python
import numpy as np
import matplotlib.pyplot as plt
month = np.array(["May", "Jun", "Jul", "Aug", "Sep"])
rain = np.array([5, 10, 14, 11, 9])
plt.plot(month, rain)
plt.xlabel("Months")
plt.ylabel("Rainfail in (mm)")
```

```python
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 4, 6, 8, 10])
y2 = np.array([1, 3, 5, 7, 9])

plt.plot(x, y1, label = "Even", linestyle = "--", color = "green", marker = "o")
plt.plot(x, y2, label = "odd", marker = "o")
plt.legend()
plt.grid()
plt.figure(figsize=(16,10))
plt.show()
```

```python
students = ["Aman", "Rahul", "Priya", "Neha", "Rohit"]
marks = [78, 92, 85, 88, 95]
plt.plot(students, marks, linestyle = "--", color = "green")
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
temperature = [18, 22, 28, 34, 39, 36]
plt.plot(months, temperature, color = "red", marker = "o", linestyle = "--", linewidth = "2")
plt.title("Monthly Temperature")
plt.xlabel("Months")
plt.ylabel("Temperatures")
plt.grid()
```
