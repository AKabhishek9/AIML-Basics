```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
y = x**2
plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o')
plt.title("Demo")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 6))
for ax in axes.flat:
    ax.plot(x, y, color='teal')
    ax.set_title('Subplot')
    ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(20, 8))
axis1 = fig.add_axes([0.1, 0.55, 0.35, 0.35])
axis1.plot(x, y, color='royalblue')
axis2 = fig.add_axes([0.55, 0.1, 0.35, 0.35])
axis2.plot(x, y, color='crimson')
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y, color='blue', marker='*')
ax.set_title('Demo Plot')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
fig.savefig("demo plot.png", dpi=200, bbox_inches='tight')
plt.show()
```
