```python
import matplotlib.pyplot as plt

students = ["Aman", "Rahul", "Priya", "Neha", "Rohit"]
marks = [78, 92, 85, 88, 95]
colors = ["red", "green", "blue", "orange", "purple"]

plt.figure(figsize=(8,5))

plt.barh(students, marks, height=0.5, color = colors)

plt.title("Student Marks")
plt.xlabel("Marks")
plt.ylabel("Students")

plt.show()
```

```python
import numpy as np
subjects = ["Math", "Physics", "Chemistry", "English"]
boys = [80, 75, 90, 85]
girls = [85, 80, 95, 88]

x = np.arange(len(subjects))
width = 0.35

plt.figure(figsize=(8,5))

plt.bar(x, boys, width=width, label="Boys")
plt.bar(x+width, girls, width=width, label="Girls")

plt.xticks(x, subjects)

plt.title("Marks Comparison")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
```

```python
months = ["Jan", "Feb", "Mar", "Apr"]
sales_2024 = [120, 150, 180, 200]
sales_2025 = [140, 160, 190, 210]
width = 0.3

plt.figure(figsize=(8, 5))
y = np.arange(len(months))
plt.bar(y-width/2, sales_2024, width = width, color = "blue", label="2024")
plt.bar(y+width/2, sales_2025, width = width, color = "orange", label="2025")
plt.xticks(y, months)
plt.title("Monthly Sales Comparison")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()

```

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr"]

boys = [25, 30, 28, 35]
girls = [20, 25, 30, 32]

plt.figure(figsize=(8,5))

plt.bar(months, boys, color="blue", label="Boys")
plt.bar(months, girls, bottom=boys, color="pink", label="Girls")

plt.title("Student Participation")
plt.xlabel("Months")
plt.ylabel("Students")
plt.legend()

plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr"]

online_sales = [50, 60, 70, 80]
offline_sales = [30, 40, 35, 45]

plt.figure(figsize=(8, 5))
plt.bar(months, online_sales, color="green", label="online_sales")
plt.bar(months, offline_sales, color="orange", bottom=online_sales, label="offline_sales")
for i in range(len(online_sales)):
    plt.text(i, online_sales[i]+2, online_sales[i])
for i in range(len(online_sales)):
    plt.text(i, offline_sales[i] + online_sales[i]+2, offline_sales[i])
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Sales Distributoion")
plt.legend()
```

```python
students = ["Aman", "Rahul", "Priya", "Neha"]
marks = [78, 92, 85, 88]

plt.figure(figsize=(8,5))

plt.bar(students, marks, color="skyblue")

for i in range(len(students)):
    plt.text(i, marks[i] - 5, str(marks[i]))

plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```
