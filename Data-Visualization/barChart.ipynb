```python
import matplotlib.pyplot as plt

students = ["Aman", "Rahul", "Priya", "Neha", "Rohit"]
marks = [78, 92, 85, 88, 95]
colors = ["red", "green", "blue", "orange", "purple"]

plt.figure(figsize=(8,5))

plt.barh(students, marks, height=0.5, color = colors)

plt.title("Student Marks")
plt.xlabel("Marks")
plt.ylabel("Students")

plt.show()
```

```python
subjects = ["Math", "Physics", "Chemistry", "English", "CS"]
marks = [88, 75, 91, 82, 96]
plt.figure(figsize=(8, 5))
plt.bar(subjects, marks, width=0.5, color = "green")
plt.title("Subjects Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
```

```python
players = ["MS Dhoni", "Rohit", "Gill", "Rahul", "Hardik"]
runs = [125, 80, 110, 65, 75]
colors = ["green", "blue", "red", "orange", "purple"]
plt.figure(figsize=(8, 5))
plt.bar(players, runs, width=0.5, color = colors)
plt.title("Player Runs")
plt.xlabel("Players")
plt.ylabel("Runs")
```

```python
import numpy as np
subjects = ["Math", "Physics", "Chemistry", "English"]
boys = [80, 75, 90, 85]
girls = [85, 80, 95, 88]

x = np.arange(len(subjects))
width = 0.35

plt.figure(figsize=(8,5))

plt.bar(x, boys, width=width, label="Boys")
plt.bar(x+width, girls, width=width, label="Girls")

plt.xticks(x, subjects)

plt.title("Marks Comparison")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
```

```python
months = ["Jan", "Feb", "Mar", "Apr"]
sales_2024 = [120, 150, 180, 200]
sales_2025 = [140, 160, 190, 210]
width = 0.3

plt.figure(figsize=(8, 5))
y = np.arange(len(months))
plt.bar(y-width/2, sales_2024, width = width, color = "blue", label="2024")
plt.bar(y+width/2, sales_2025, width = width, color = "orange", label="2025")
plt.xticks(y, months)
plt.title("Monthly Sales Comparison")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()

```
